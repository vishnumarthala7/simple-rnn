```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=['the glass of milk',
      'the glass of juice',
      'the cup of tea',
      'understand the meaning of words',
      'I am good boy',
      'i am  a good  developer',
      'Your videos are good'
      ]
```

```python
sent
```

```python
# define the voculabary size
voc_size=10000-
```

```python
#one hot representation
one_hot_rep=[one_hot(words,voc_size)for words in sent]
one_hot_rep

```

```python
# word embeding representation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences # type: ignore
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sent_length=8
embeded_docs=pad_sequences(one_hot_rep,padding='pre',maxlen=sent_length)
embeded_docs
```

```python
#feature repressentation
dim=10
```

```python
model=Sequential()
model.add(Embedding(input_dim=voc_size,output_dim=dim,input_length=sent_length))
model.compile(optimizer='adam',loss='categorical_crossentropy')
```

```python
model.summary()
```

```python
model.predict(embeded_docs)
```
